# preprocessing data set

```python
# all NaN value are fill up and 30% NaN value columns are removing
def replacer(a):
    import pandas as pd
    Q=pd.DataFrame(a.isna().sum(),columns=["ct"])
    for i in Q[Q.ct>0].index:
        if (a[i].dtype=="object"):
            x=a[i].mode()[0]
            a[i]=a[i].fillna(x)
        else:
            x=a[i].mean()
            a[i]=a[i].fillna(x)
```

```python
replacer(a)
```

```python
a.info()
```

# dividing the data catagorical and continious value

```python
cat=[]
con=[]
```

```python
# cat value and con value sepereable
for i in a.columns:
    if (a[i].dtype=="object"):
        cat.append(i)
    else:
        con.append(i)
```

```python
X1=pd.get_dummies(a[cat])
```

```python
X2=a[con]
```

```python
X3=X2.join(X1) # join cat and con columns
```

```python
X3
```

```python
Y=pd.DataFrame(A["Status"])
```

# splitting the data two way train and test set

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X3,Y) # split the data training and testing set 
```

```python
from sklearn.neighbors import KNeighborsClassifier# model import
```

```python
ll=KNeighborsClassifier(n_neighbors=20)
```

```python
tt=ll.fit(xtrain,ytrain)
```

```python
tr=tt.predict(xtrain)
```

```python
tr
```

```python
ts=tt.predict(xtest)
```
